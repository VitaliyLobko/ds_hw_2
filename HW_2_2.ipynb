```python
#Створіть нову таблицю python_data, в якій будуть тільки рядки зі спеціалістами, які вказали мову програмування Python
python_data = c_df[c_df['Язык.программирования'] == 'Python']
python_data
```

```python
#Визначте розмір таблиці python_data за допомогою методу shape
python_data.shape
```

```python
#Використовуючи метод groupby, виконайте групування за стовпчиком "Посада
python_data_group = python_data.groupby('Должность')
python_data_group.first()
```

```python
#Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"
python_data_groupby = python_data.groupby(['Должность'])[['Зарплата.в.месяц']].agg(['min', 'max'])
python_data_groupby
```

```python
x = python_data_groupby['Зарплата.в.месяц'].index.tolist()
y_min = python_data_groupby['Зарплата.в.месяц']['min'].values.tolist()
y_max = python_data_groupby['Зарплата.в.месяц']['max'].values.tolist()

width = 0.25
multiplier = 0

jobs = x

wage_means = {
    'min': y_min,
     'max': y_max
}


x = np.arange(len(jobs))

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in wage_means.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=6)
    multiplier += 1

ax.set_ylabel('$')
ax.set_title('заробітна плата')
ax.set_xticks(x + width, jobs)
ax.legend(loc='upper left', ncols=3)
plt.rcParams ['figure.figsize'] = [12, 8]

# fig, ax = plt.subplots()
# ax.set_title("заробітна плата")
# ax.barh(x, y_min, color="red", label="")
```

```python
#Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць. Використовуйте її для методу apply та створіть новий стовпчик "avg"
def fill_avg_salary(python_data):
  new_data = python_data.groupby(['Должность'])[['Зарплата.в.месяц']].agg(['max'], ['min'])
  return (new_data.sum(axis=1)/2)


mean = fill_avg_salary(python_data)
mean

python_data_groupby['avg'] = mean.apply(lambda x: x)
python_data_groupby
```

```python
#Створіть описову статистику за допомогою методу describe для нового стовпчика.
python_data_groupby['avg'].describe()
```
